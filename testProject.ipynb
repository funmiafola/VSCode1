```python
import pandas as pd
# Load the Bank Rate CSV
BankRate = pd.read_csv(r"\\talktalk\redirectedfolders\F.Afolabi\Documents\VSCode1\Bank_rate_england.csv")

# Normalise column names
BankRate.columns = BankRate.columns.str.strip().str.lower()

# Rename the columns
BankRate.rename(columns = {'date':'Date', 'rate':'Rate'}, inplace=True)

BankRate['Date'] = pd.to_datetime(
    BankRate['Date'], errors='coerce')

# 4. Drop rows with no date 
BankRate = BankRate.dropna(subset=['Date'])

# Check the data
print(BankRate.head())
print(BankRate.info())

#  Sorting  by date
BankRate = BankRate.sort_values('Date')


# Filtering the data between 2019 and 2025 for Bank Rate
BankRate = BankRate[(BankRate['Date'] >= '2019-01-01') & (BankRate['Date'] <='2025-12-31')]
print(BankRate.head())
BankRate.info()

#Resample to quarterly to aligh with CPI data
BankRate_quarterly = BankRate.set_index('Date').resample('QS').mean().reset_index()
print(BankRate_quarterly.head())


# filtering the data between 2019 and 2025 for CPI
cpi = cpi[(cpi['Date'] >= '2019-01-01') & (cpi['Date'] <='2025-12-31')]
print(cpi.head())
cpi.info()

```

```python
#Merge the CPI and Bank Rate 
merged_data = pd.merge(cpi, BankRate_quarterly, on='Date', how='inner')
print(merged_data.head())
print(merged_data.tail())
```

```python
#Remove duplicates
merged_data = merged_data.drop_duplicates()
print(merged_data.info())
print(merged_data.head())
```
