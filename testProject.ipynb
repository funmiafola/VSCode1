```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load your CSV
boe_rate = pd.read_csv("Bank_of_England_Base_Rate.csv")

# Step 2: Inspect the data
print(boe_rate.head())
print(boe_rate.info())
```

```python
print(f"Number of columns: {boe_rate.shape[1]}")
```

```python
boe_rate = boe_rate.rename(columns={
    boe_rate.columns[1]: "2yr_60_LTV",
    boe_rate.columns[2]: "2yr_75_LTV",
    boe_rate.columns[3]: "2yr_85_LTV",
    boe_rate.columns[4]: "2yr_90_LTV",
    boe_rate.columns[5]: "2yr_95_LTV",
})
```

```python
boe_rate['Date'] = pd.to_datetime(boe_rate['Date'], dayfirst=True)
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=boe_rate, x='Date', y='2yr_60_LTV', label='2yr 60% LTV')
sns.lineplot(data=boe_rate, x='Date', y='2yr_75_LTV', label='2yr 75% LTV')
sns.lineplot(data=boe_rate, x='Date', y='2yr_85_LTV', label='2yr 85% LTV')
plt.title("UK 2-Year Fixed Rate Mortgages")
plt.xlabel("Date")
plt.ylabel("Interest Rate (%)")
plt.legend()
plt.show()


```
